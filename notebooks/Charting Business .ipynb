#### We are converting the values of our address colum to latitude and longitude

```python
locator = Nominatim(user_agent= 'myGeocoder')
```

```python
from geopy.extra.rate_limiter import RateLimiter

# 1 - conveneint function to delay between geocoding calls
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
# 2- - create location column
hate_form['location'] = hate_form['Address'].apply(geocode)
# 3 - create longitude, laatitude and altitude from location column (returns tuple)
hate_form['point'] = hate_form['location'].apply(lambda loc: tuple(loc.point) if loc else None)
# 4 - split point column into latitude, longitude and altitude columns
hate_form[['latitude', 'longitude', 'altitude']] = pd.DataFrame(hate_form['point'].tolist(), index=hate_form.index)
```

```python
hate_form
```

#### We are graphing the data to show the business and the one's that need pandemic assistance that are Asian. 

```python
#icon_url = './img/covid-red.png'  # to change icon
bronx = [40.8448, -73.8648]
popup = folium.Popup(min_width=5550)
lat = hate_form.latitude.values
lon = hate_form.longitude.values
zipcode = hate_form['Zip Code'].values
business_name = hate_form['Business Name'].values   #area
assistance = hate_form['Pandemic Assistance'].values  #people positive
nl = '\n'
points = list(zip(lat, lon, zipcode, business_name, assistance))
m = folium.Map(location=bronx, zoom_start=11, tiles="openstreetmap")
for lat, lon, zipcode, business_name, assistance in points:
    folium.Marker(
        [lat,
            lon], popup=f"<p>Zipcode: {zipcode}{nl} Buisness Name: {business_name}{nl} Pandemic Assitance:{assistance}{nl}</p>"
    ).add_to(m)
m.save("folium.html")
m
```
