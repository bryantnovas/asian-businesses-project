```python
search = gn.search('Anti-Asian Hate Crimes', when = '16m')
```

```python
 search 
```

#### Parse through the entries because we want to specifically identify the titles of the loop.

```python
search['entries']
```

#### We are storing our entries into a string because our information is in json format. Each line will be represented into a new line and printed within the titles only of our search

```python
words = ""
for entry in search['entries']:
    words += entry["title"]+ '\n'
print(words)
```

```python
str(words)
```

```python
links = ""
for entry in search['entries']:
    links += entry["link"]+ '\n'
print(links)
```

#### We are now searching through the urls for the keywords that we identifed and returning the new value within the dictionary of the count of how many times these words appeared within the articles. 

```python
for url in list_of_urls:
    text = requests.get(url).text
    for k,_ in x.items():
        x[k] += text.count(k)
        
x
```

### We know turn this dictionary into a dataframe to graph it, and see the occurence of the words.

```python
# turn this dictionary into a dataframe and graph it 
occurence = pd.DataFrame(list(x.items()),columns = ['Common Words','Count']) 
```

```python
occurence
```

#### The .sum function allows us to get the total for each common word and the count  .

```python
occurence.groupby('Common Words')['Count'].sum()
```
